### Imports

```python
from pathlib import Path
import pandas as pd
```

### Constants

```python
base_dir = Path("../").resolve()
file     = base_dir/'results/final/wide_df.csv'
out_dir  = base_dir/'results/final/'
```

### Main

```python
df =pd.read_csv(file)
df.head()
```

```python
df.shape
```

```python
depth_cols  = df.columns[3:]
df['mean'] =df[depth_cols].mean(axis=1, skipna=True)
df
```

```python
df['mean'].quantile([0.40,0.40,0.50,0.60,0.70,0.80,0.90,0.99])
```

```python
df["frac_gt_50"] = df[depth_cols].gt(50).sum(axis=1) / len(depth_cols)
df["frac_gt_50"].quantile([0.20,0.30,0.40,0.45,0.50,0.60,0.70,0.80,0.90,0.99])
```

```python
filtered = df[df["frac_gt_50"] >= 0.8].copy()
```

```python
filtered
```

```python
filtered.describe()
```

```python
filtered = filtered[['chr','start','end']]
filtered.to_csv(out_dir/'autosome_exons_filtered.bed',sep="\t",index=False)
```
